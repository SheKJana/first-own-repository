## Einleitung

Ernährungsbedingte Krankheiten wie Typ-2-Diabetes, Bluthochdruck und Adipositas nehmen weltweit zu. Gleichzeitig bleibt der Zugang zu individueller, präventiv wirksamer Ernährungsberatung beschränkt: Klassische Beratungen sind kostenintensiv und nicht skalierbar, während pauschale Ernährungsempfehlungen oft an den tatsächlichen Bedürfnissen vorbeigehen. Dank moderner Datenanalyse und Machine Learning eröffnen sich heute jedoch neue Möglichkeiten, personalisierte Ernährungsempfehlungen datenbasiert, automatisiert und massentauglich zu gestalten.
Im Rahmen dieser Fallstudie simulieren wir die Entwicklung eines fiktiven digitalen Empfehlungstools namens „NutriDecide“. Dieses Tool soll auf Basis realer Essgewohnheiten, Nährstoffverteilungen und Gesundheitsdaten konkrete Empfehlungen für eine ausgewogene, gesundheitsfördernde Ernährung ableiten.

Wir orientieren uns dabei am **CPA-Framework _"From Data to Decisions"_** und fokussieren uns insbesondere auf folgende Schritte:

1. Zieldefinition und Informationsbedarf  
2. Datenerhebung und -integration  
3. Datenanalyse  
4. Kommunikation der Erkenntnisse für Entscheider

Zwar existieren bereits zahlreiche virtuelle Ernährungstools und -Apps auf dem Markt, jedoch fokussieren sich diese häufig primär auf Gewichtsverlust oder allgemeine Kalorienziele. Eine fundierte, auf wissenschaftlichen Empfehlungen basierende Risikoabschätzung hinsichtlich chronischer Erkrankungen fehlt meist.

NutriDecide setzt genau hier an und verfolgt einen präventiven, individualisierten Ansatz mit klarer Zielsetzung: Die Identifikation gesundheitlich bedenklicher Ernährungsmuster und die Ableitung konkreter Verbesserungsvorschläge auf Basis anerkannter Referenzwerte (z.B. BLV).
Das Tool adressiert den steigenden Bedarf an Gesundheitsprävention und soll potenziell über Krankenkassen angeboten werden. Diese können damit ihren Versicherten einen kostenlosen Zugang zu individuellen Ernährungsempfehlungen bieten – und gleichzeitig durch gezielte Prävention langfristig die Gesundheitskosten senken. „NutriDecide“ positioniert sich damit als innovatives Bindeglied zwischen Versicherer, Nutzer:in und Gesundheitsförderung.

## 1 Zieldefinition und Informationsbedarf
Damit ein digitales Ernährungstool wie NutriDecide überhaupt sinnvoll entwickelt und eingesetzt werden kann, braucht es zuerst eine solide Datengrundlage. Dazu gehört nicht nur, zu wissen, was Menschen essen, sondern auch, wie sich diese Essgewohnheiten auf ihre Gesundheit auswirken – und welche Empfehlungen daraus abgeleitet werden können.

Im Sinne des CPA-Frameworks ist es deshalb entscheidend, sich zuerst die richtigen Fragen zu stellen, bevor man technische Lösungen entwickelt.
Nur wenn klar ist, welche Informationen wirklich gebraucht werden, lässt sich ein Tool gestalten, das am Ende für Nutzer:innen oder Organisationen einen echten Mehrwert bringt.

Genau deshalb haben wir vier zentrale Key Analytics Questions (KAQs) formuliert, die unsere Analyse leiten. Sie helfen uns zu verstehen, welche Daten wir brauchen, wie wir sie kombinieren müssen – und was wir daraus lernen wollen.

|------------:|----------------------------|-----------|--------------------------|
| 1 | In welchen Ernährungsgewohnheiten liegen potenziell unausgewogene Nährstoffzufuhr vor?

##  2 Daten

Für die Beantwortung unsere Fragen und dem entwickeln des Tools benötigen, wir Daten welche sich dafür eignen.

Bei der Auswahl unserer Datenquellen haben wir uns an folgenden Kriterien orientiert:

- Die Daten mussten die für unsere Fragestellungen relevanten Informationen enthalten, insbesondere Angaben zur individuellen Nährstoffzufuhr und zum Gesundheitszustand.

- Die Daten sollten eine ausreichende Anzahl von Beobachtungen aufweisen, um statistische Analysen mit gewisser Aussagekraft durchführen zu können.

- Die Struktur der Datensätze sollte grundsätzlich verständlich und weiterverarbeitbar sein – auch wenn wir eine eigene Bereinigung und Aufbereitung vornehmen würden.


 **Verwendete Datensätze**

Für die vorliegende Fallstudie war von Beginn an klar, dass ein qualitativ hochwertiger, personenbasierter Datensatz erforderlich ist, der sowohl Angaben zur individuellen Ernährung als auch zu Gesundheitszuständen umfasst. Die Suche nach einem geeigneten Datensatz erfolgte gezielt über die Plattform Kaggle, die eine breite Sammlung öffentlich zugänglicher, teils internationaler Forschungsdatensätze bietet (Kaggle, o. J.).

Ursprünglich war die Analyse mit einem Datensatz geplant, der detaillierte Informationen zu individuellen Mahlzeiten und Krankheitsbildern enthielt. Bei näherer Prüfung zeigte sich jedoch eine methodische Schwäche: 100 % der erfassten Personen wiesen mindestens eine diagnostizierte Krankheit auf. Somit fehlte eine gesunde Vergleichsgruppe, was eine differenzierte Auswertung zur Prävention und zum Risikoausgleich verunmöglichte.

1. ** NHANES Datensatz **
Aus diesem Grund fiel die Entscheidung auf den Datensatz der National Health and Nutrition Examination Survey (NHANES), der durch die Centers for Disease Control and Prevention (CDC) erhoben und auf Kaggle veröffentlicht wird (CDC, o. J.). NHANES kombiniert Gesundheitsinterviews mit klinischen Untersuchungen und Ernährungsprotokollen und zählt international zu den etabliertesten, wissenschaftlich anerkannten Studien zur Gesundheits- und Ernährungslage der Allgemeinbevölkerung.

Die verwendete Datengrundlage umfasst insgesamt 9’813 Personen und ermöglicht es, sowohl gesunde als auch erkrankte Teilnehmende in die Analyse einzubeziehen. Dadurch konnten relevante Vergleichsgruppen konstruiert und individuelle Abweichungen zur empfohlenen Nährstoffzufuhr datengestützt bewertet werden.



2. **Food and Nutrition Dataset**
Der „Food and Nutrition Dataset“ (Utsavdey, o. J.) enthält eine umfangreiche Liste von Lebensmitteln mit detaillierten Angaben zu Kalorien, Protein, Kohlenhydraten und Fett. Die strukturierte und standardisierte Aufbereitung ermöglicht eine zuverlässige Nährwertanalyse auf Einzelproduktebene.

Diese Datenbasis ist besonders nützlich, um Nährstoffziele präzise mit geeigneten Lebensmitteln zu verknüpfen. Sie bildet die Grundlage für bedarfsorientierte Zuordnungen und die Entwicklung tagesbasierter Ernährungsvorschläge.

3. **BLV Referenzwerte**  
Die Referenztabelle „Nutrient Requirements CH – 2022“ enthält offizielle Schweizer Empfehlungen zur täglichen Zufuhr von Energie und Makronährstoffen, differenziert nach Alter, Geschlecht und physiologischen Bedingungen. Sie ermöglicht dadurch eine individuelle Bewertung der Ernährung im Verhältnis zu wissenschaftlich anerkannten Sollwerten und bildet die Grundlage für eine differenzierte und bedarfsgerechte Analyse des Nährstoffhaushalts.

##  3 Datenanalyse

Um eine Übersicht über die Daten zu erhalten, lag der Fokus:

** 3.1 In welchen Ernährungsgewohnheiten liegen potenziell unausgewogene Nährstoffzufuhr vor?**

Bevor wir die Daten einem vordefinierten Kategorienschema unterwarfen, haben wir exploratives Clustering auf Basis der individuellen Nährstoffaufnahme durchgeführt. Berücksichtigt wurden dabei zentrale Makro- und Mikronährstoffe wie Kalorien, Eiweiss, Gesamtfett, gesättigte Fettsäuren, Ballaststoffe und Natrium. Ziel war es, natürliche Ernährungsmuster unabhängig von vorgegebenen Referenzwerten zu identifizieren. Die resultierenden Cluster wiesen jedoch eine geringe Qualität auf: Der Silhouette Score lag bei 0.13 und damit deutlich unter dem empfohlenen Schwellenwert von 0.2. Dies deutet auf eine geringe Trennschärfe und eine hohe Überlappung zwischen den Gruppen hin – ein Hinweis darauf, dass sich auf Basis der betrachteten Nährstoffe keine klar unterscheidbaren Muster herausbilden.
Ein möglicher Grund für das Scheitern des explorativen Clusterings liegt darin, dass absolute Nährstoffmengen allein keine zuverlässige Grundlage für die Beurteilung von Ernährungsmustern darstellen. Eine Ernährung ist nur im Kontext der individuellen physiologischen Voraussetzungen bewertbar – insbesondere Alter, Körpergrösse, Geschlecht und Aktivitätsniveau beeinflussen stark, was als angemessen oder unausgewogen gilt.
Da diese relativen Anforderungen in den Rohdaten nicht berücksichtigt wurden, führt eine rein datengetriebene Gruppierung auf Basis der aufgenommenen Mengen zu einer inhaltlichen Verzerrung: Individuen mit objektiv sehr unterschiedlichen Versorgungsniveaus können aufgrund ähnlicher Absolutwerte in dieselbe Gruppe fallen, während tatsächlich gleichversorgte Personen auf mehrere Cluster verteilt werden.
Diese fehlende Kontextualisierung der Werte dürfte wesentlich dazu beigetragen haben, dass sich keine klar trennbaren Muster herausbilden konnten.

Umkehrung des Ansatzes: Klassifizierung nach unausgewogen und ausgewogen
Da das datengetriebene Clustering auf Basis der Rohdaten keine verwertbaren Ergebnisse lieferte, haben wir das methodische Vorgehen umgekehrt: Bedeutet statt zuerst Ernährungsmuster aus den Rohdaten zu eruieren, widmen wir uns zuerst der Klassifizierung ausgewogene oder unausgewogen.
Definition von unausgewogener Ernährung
Eine Ernährung gilt dann als unausgewogen, wenn sie langfristig nicht im Einklang mit den empfohlenen Richtwerten für Energie- und Nährstoffzufuhr steht. Dies kann zu Mangelzuständen, Überversorgung oder ernährungsbedingten Erkrankungen führen.
Zuerst wurde für jede Person individuell beurteilt, wie stark ihre Nährstoffaufnahme von den wissenschaftlich empfohlenen Referenzwerten abweicht. Zur Bewertung der Angemessenheit stützten wir uns auf die offiziellen Empfehlungen des Bundesamts für Lebensmittelsicherheit und Veterinärwesen (BLV) sowie der Weltgesundheitsorganisation (WHO). Diese enthalten sowohl absolute Grenzwerte (z. B. maximal 2000 mg Natrium pro Tag) als auch individuell berechnete Zielwerte, etwa:
•	Kalorienbedarf: berechnet aus *1Grundumsatz × *2Aktivitätsfaktor (PAL), unter Berücksichtigung von Alter, Geschlecht und Bewegung
•	Protein: 0.8 g pro Kilogramm Körpergewicht
•	Ballaststoffe: 14 g pro 1000 kcal Energiezufuhr
•	Fett, Zucker: empfohlen als Anteil des täglichen Energiebedarfs
*1Der Grundumsatz wird der BVL Tabelle entnommen
*2 PAL ist ein Tagesdurchschnitt aller Aktivitäten von Beruflicheaktivitäten bis Freizeitaktivitäten. Die Referenzwerte haben wir aus https://www.fao.org/3/y5686e/y5686e00.htm

wir haben alle Personen herausgefiltert, für die vollständige Angaben zu Ernährung, Körperdaten und mindestens eine Angabe zum Aktivitätsverhalten vorlagen. Personen ohne jegliche Bewegungseinträge wurden ausgeschlossen, da für sie keine PAL-Klassifikation möglich war. Für alle übrigen Teilnehmenden wurden fehlende Werte in den Aktivitätsfeldern (PAD615–PAD675) mit 0 ergänzt – da hier von keiner Aktivität in dem jeweiligen Bereich ausgegangen werden kann. Auf Basis dieser bereinigten Daten wurde der PAL-Wert jeder Person bestimmt, um die individuell empfohlene Energie- und Nährstoffzufuhr zu berechnen. Anschliessend ermittelten wir für jede Person die prozentuale Abweichung bei Kalorien, Protein, Ballaststoffen und Natrium im Vergleich zu ihrem individuellen Sollwert.

Diese normierten Abweichungswerte wurden standardisiert und mithilfe von k-means (k=3) in Ernährungsmuster gruppiert. Die Clusterqualität mit einem Silhouette Score von 0.36 ist als akzeptabel einzustufen. Inhaltlich zeigen sich drei klar unterscheidbare Muster: (1) Unterversorgung bei Protein und Ballaststoffen, (2) Überversorgung bei Kalorien und Natrium, (3) ausgeglicheneres Profil mit leichtem Energieüberschuss. Diese Segmentierung erlaubt eine differenzierte Identifikation potenziell kritischer Ernährungstypen.
